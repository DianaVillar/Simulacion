```python
import scipy.stats as ss
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
import scipy.stats as stats

df=20 #grados de libertad
chi2= stats.chi(df)
x= np.linspace(chi2.ppf(0.01),
                          chi2.ppf(0.99),100)
fp= chi2.pdf(x)
plt.plot(x,fp, "g")
plt.title("Distribucion chi cuadrada")
plt.ylabel("probabilidad")
plt.xlabel("valores")
plt.show()

#histograma
aleatorios= chi2.rvs(1000)
a,b,c=plt.hist(aleatorios,20,color="red")
plt.title("Histograma de la distribucion chi cuadrada")
plt.ylabel("frecuencias")
plt.xlabel("valores")
plt.show()
```
