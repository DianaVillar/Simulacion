```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def prueba_num_al(random_numbers, num_bins):
    # Paso 1: Calcular la frecuencia observada
    frecuencia_observada, bins = np.histogram(random_numbers, bins=num_bins)

    # Paso 2: Calcular la frecuencia esperada
    frecuencia_esperada = len(random_numbers) / num_bins

    # Paso 3: Calcular el estadístico Chi-Cuadrado
    estadistico_chi = np.sum((frecuencia_observada - frecuencia_esperada) ** 2 / frecuencia_esperada)

    # Paso 4: Calcular los grados de libertad
    grados_libertad = num_bins - 1

    # Paso 5: Calcular el valor p
    p_value = 1 - stats.chi2.cdf(estadistico_chi, grados_libertad)

    # Mostrar los resultados
    print(f"Estadistico Chi cuadrada: {estadistico_chi}")
    print(f"Grados de Libertad: {grados_libertad}")
    print(f"P-Value: {p_value}")

    # Graficar la distribución observada y esperada
    plt.hist(random_numbers, bins=num_bins, alpha=0.5, label='Observaciones')
    plt.axhline(y=frecuencia_esperada, color='r', linestyle='dashed', linewidth=1, label='Experado')
    plt.legend(loc='upper right')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    plt.title('Prueba de Chi cuadrada para numeros aleatorios')
    plt.show()

# Generar números aleatorios
num_samples = 1000
random_numbers = np.random.uniform(0, 1, num_samples)

# Ejecutar la Prueba de Forma Chi-Cuadrada
num_bins = 10  # Número de intervalos
prueba_num_al(random_numbers, num_bins)
```
